```python
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties

font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
student1 = {'國文':78, '英文': 80, '數學':75, '歷史': 85, '探索':92}
student2 = {'國文':90, '英文': 75, '數學':98, '歷史': 65, '探索':75}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.plot(subjects, scores1, label='stu1', marker='.')
axes.plot(subjects, scores2, label='stu2', marker='s')
axes.set_xticklabels(subjects,fontproperties=font)
axes.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei

#codespace
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
#windows
#font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#============#
#mac
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)

student1 = {'國文':78, '英文': 80, '數學':75, '歷史': 85, '探索':92}
student2 = {'國文':90, '英文': 75, '數學':98, '歷史': 65, '探索':75}

subjects=list(student1.keys())
scores1=list(student1.values())
scores2=list(student2.values())

fig=plt.figure(figsize=(8, 5),dpi=250,facecolor='#cccccc')
axes=fig.add_subplot(1,1,1)
axes.plot(subjects,scores1,'r^-.',label='stu1')
axes.plot(subjects,scores2,'go--',label='stu2')
axes.set_xticklabels(subjects,fontproperties=font)
axes.legend()
axes.set_xlabel("X",fontproperties=font)
axes.set_ylabel("Y",fontproperties=font)

plt.show()
a=axes.get_xticks()
for i in range (len(a)):
    axes.annotate(str(student1[i],(a[i]+0.1,scores1[i])))
    axes.annotate(str(student2[i],(a[i]+0.1,scores1[i])))
    
```
