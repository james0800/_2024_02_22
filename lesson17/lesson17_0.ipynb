```python
import pandas as pd
data=pd.read_csv("world.csv")
data=data.reindex(columns=["國家","日期",'總確診數','新增確診數','新增死亡數'])

display(data[(data["國家"]=="台灣") | (data["國家"]=="日本") ])

```

```python
data2=data.query("國家=='日本' or  國家=='台灣'")
display(data2)
```

```python
data3=data2.query("國家=='台灣' and  新增確診數==0")
display(data3)
data3.to_excel("台灣0.xlsx")
```
