```python
sc=np.random.randint(50,101,size=(50,5))
sc=pd.DataFrame(sc,columns=['國文', '英文', '數學', '地理', '歷史'])
sc2=sc
sm=sc.sum(axis=1)
av=sc.mean(axis=1)
sc["sum"]=sm

sc["ave"]=av
sc
```

```python
from pandas import Series
def f(s:Series):
    av=s.mean()
    if av<70:
        return "F"
    elif av<80:
        return "C"
    elif av<90:
        return "B"
    else:
        return "A"
sc2["不及格科目"]=sc2[['國文', '英文', '數學', '地理', '歷史']].apply(f,axis=1)
#display(sc2)
display(sc2)
```

```python
#國文	英文	數學	地理	歷史 不及格科目
def fail(s:Series):
    sub=['國文', '英文', '數學', '地理', '歷史']
    n=0
    for i in sub:
       
        if s[i]<80:
           n+=1

    return n
sc2["不及格科目"]=sc2[['國文', '英文', '數學', '地理', '歷史']].apply(fail,axis=1)
display(sc2)   
    
        
        
```
