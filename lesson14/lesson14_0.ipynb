```python
import numpy as np
import matplotlib.pyplot as plt
arr1=np.array([10,20,30])
arr2=np.array([1,2,3])
axes=plt.subplot()
print(arr1)
print(arr1*2)
print(arr1*arr2)
print(arr1.shape)
b=np.linspace(2.0,3.0,num=100)
y1=b*2+1
y2=b**2
axes.plot(b,y1)
axes.plot(b,y2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,num=50)

y1 = x * 2 + 1
y2 = x ** 2

figure, axes = plt.subplots()
axes.plot(x,y1)
axes.plot(x,y2,color='red',linewidth=1.0, linestyle='--')
axes.set_xlim((-1,1))
axes.set_ylim((-2,2))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
axes=plt.subplot()
x=np.arange(0,2.5,0.1)
y1=np.sin(math.pi*x)
y2=np.sin(math.pi*x+10)
axes.plot(x,y1)

axes.plot(x,y2)
plt.show()

```
