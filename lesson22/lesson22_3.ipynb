```python
import requests
import pandas as pd
url = 'https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000'
re=requests.get(url)
youe=re.json()
youe
```

```python

youbike_df = pd.DataFrame(youe,
             columns=['sna', 'tot', 'sbi', 'sarea', 'ar', 'bemp'])
youbike_df.columns = ['站名','數量','可借','行政區','地址','可還']
youbike_df
```

```python
youbike_df[["數量","可借","可還"]]=youbike_df[["數量","可借","可還"]].astype(int)
youbike_df.groupby("行政區")[["數量","可借","可還"]].sum()
```

```python
youbike_df.groupby('行政區')[['數量','可借','可還']].agg([('總量','sum'),('最多','max'),('最少','min')])
```
