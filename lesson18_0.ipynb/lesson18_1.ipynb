```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('各鄉鎮市區人口密度.csv')

df1.columns = df1.iloc[0]

df2 = df1.drop(index=0)

df3 = df2.dropna()

df4 = df3.set_index(['區域別'])
df5 = df4.drop(columns=['統計年'])
df5.info()
```

```python
def to_int(v:str)->int:
    try:
        return int(v)
    except:
        return 0
df5["年底人口數"]=df5["年底人口數"].map(to_int)
df5
```
