```python
import random
import pyinputplus as pyip
from pprint import pprint
import csv
from csv import DictWriter
import tools
import lesson11_0
num:int = pyip.inputInt('請輸入指定人數(1~50):', min=1, max=50)
names:list[str] =tools.get_names(num)
items:list[dict[str:any]] = lesson11_0.generateData(names)
with open('students.csv',mode='w',encoding='utf-8',newline='') as csv_file:
    writer:DictWriter = csv.DictWriter(csv_file,['name','height','weight'])
    writer.writeheader()
    writer.writerows(items)
print("儲檔完成")
```

```python
from requests import Response
import requests
import csv
def main()->int:
    link = 'https://raw.githubusercontent.com/roberthsu2003/__2024_02_20_2_4__/main/lesson11/students.csv'
    try :
        down:Response=requests.get(link)
        
    except:
        print("Erro")
        return 1
    file:str=down.text
    b:StringIO=StringIO(file)
    cs=csv.DictReader(b)
    data:list[dict]=list(cs)
    for i in data:
        print(i)
    return 0
main()
```
